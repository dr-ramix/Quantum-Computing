# Task 4-3 (Teleportation-Simulation)
Recreating the circuit for teleportation from the lecture into the simulator.

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit_aer.backends.aer_simulator import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
def quantum_circuit():
    quantum_register = QuantumRegister(3)
    classical_register = ClassicalRegister(1)
    quantum_circuit = QuantumCircuit(quantum_register,classical_register)
    #Initial Bell-State |B00>
    quantum_circuit.h(1)
    quantum_circuit.cx(2,1)
    #Ψ0
    quantum_circuit.cx(0,1)
    #Ψ1
    quantum_circuit.h(0)
    #Ψ2
    quantum_circuit.measure(0, classical_register)
    quantum_circuit.measure(1, classical_register)
    #Ψ3
    quantum_circuit.cx(1,2)
    #Ψ4
    quantum_circuit.cz(0,2)
    #Ψ5
    quantum_circuit.measure(2, classical_register)
    return quantum_circuit

```

```python
qc = quantum_circuit()
qc.draw("mpl")
```

```python
simulator = AerSimulator()
result = simulator.run(qc).result()
plot_histogram(result.get_counts(0))
```
